```python
import numpy as np
def relu(x):
    return np.maximum(0, x)
def initialize_network():
    input_size = 5
    hidden_layer1_size = 3
    hidden_layer2_size = 2
    hidden_layer3_size = 3
    output_size = 1
    
    #Weights for each layer
    weights_layer1 = np.random.rand(input_size, hidden_layer1_size)
    weights_layer2 = np.random.rand(hidden_layer1_size, hidden_layer2_size)
    weights_layer3 = np.random.rand(hidden_layer2_size, hidden_layer3_size)
    weights_output = np.random.rand(hidden_layer3_size, output_size)
    
    #Biases for each layer
    biases_layer1 = np.random.rand(1,hidden_layer1_size)
    biases_layer2 = np.random.rand(1,hidden_layer2_size)
    biases_layer3 = np.random.rand(1,hidden_layer3_size)
    biases_output = np.random.rand(1,output_size)
    
    # Print the architecture
    print("Neural Network Architecture: ")
    print(f"Input Layer: {input_size} nodes")
    print(f"Hidden Layer 1: {hidden_layer1_size} nodes")
    print(f"Hidden Layer 2: {hidden_layer2_size} nodes")
    print(f"Hidden Layer 3: {hidden_layer3_size} nodes")
    print(f"Output Layer: {output_size} nodes")
    return (weights_layer1, biases_layer1, weights_layer2, biases_layer2,
            weights_layer3, biases_layer3, weights_output, biases_output)
    
    #Weights and biases Printing
    print("Weights and Biases:")
    print("Weights Layer 1:\n", weights_layer1)
    print("Biases Layer 1:\n", biases_layer1)
    print("Weights Layer 2:\n", weights_layer2)
    print("Biases Layer 2:\n", biases_layer2)
    print("Weights Layer 3:\n", weights_layer3)
    print("Biases Layer 3:\n", biases_layer3)
    print("Weights Output:\n", weights_output)
    print("Biases Output:\n", biases_output)
    
    #Compute activations for each node in the first hidden layer
    input_layer = np.random.rand(1, input_size)
    activations_hidden1 = relu(np.dot(input_layer, weights_layer1) + biases_layer1)
    
    #Compute activations for each node in the second hidden layer
    activations_hidden2 = relu(np.dot(activations_hidden1, weights_layer2) + biases_layer2)
    
    #Compute activations for each node in the third hidden layer
    activations_hidden3 = relu(np.dot(activations_hidden2, weights_layer3) + biases_layer3)
    
    #Compute output layer activation
    output_activation = np.dot(activations_hidden3, weights_output) + biases_output
    
    print("Activations of Hidden Layer 1:\n", activations_hidden1)
    print("Activations of Hidden Layer 2:\n", activations_hidden2)
    print("Activations of Hidden Layer 3:\n", activations_hidden3)
    print("Output Activation:\n", output_activation)
```
